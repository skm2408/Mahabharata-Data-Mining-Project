```python
rows=[]
```

```python
lst=[0,78,120,76,68,69,3]
for num in range(1,lst[6]):
        c+=1
        URL = "https://www.valmiki.iitk.ac.in/sloka?field_kanda_tid=6&language=dv&field_sarga_value="+str(num)
        page = requests.get(URL)
        soup = BeautifulSoup(page.content, "html.parser")
        
        phrases=soup.find_all('div',class_="views-field views-field-field-explanation")
#         print(type(phrases))
        phrases=phrases[:-1]
        for row in phrases:
            for k in row.find_all('div',class_="field-content"):
                st = (k.contents)
                for k in st:
                    if type(k) is bs4.element.NavigableString:
                        rows.append([str(k).strip()])
                    if len(st)==1:
                        rows.append([(str(k).replace('<br>','').replace('</br>','')).strip()])
```

```python
with open('ramayana.csv', 'w') as csvfile: 
    csvwriter = csv.writer(csvfile) 
    csvwriter.writerows(rows)
```
